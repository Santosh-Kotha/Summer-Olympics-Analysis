```python
import pandas as pd
df =  pd.read_csv('summer.csv')
df
```

### **In how many Cities Summer Olympics are held so far**

```python
ans = len(df['City'].unique())
ans
```

### **In how many countries summer olympics are held?**

```python
ans = len(df['Country'].unique())
ans
```

### **Which Sport has more no of Gold medals(TOP 5)**

```python
sport_list = df.Sport.unique()
data = []
for sport in sport_list:
  filt = (df['Sport'] == sport) & (df['Medal'] == 'Gold')
  data.append((sport,len(df[filt])))
data = pd.DataFrame(data , columns = ['Sport','No of Gold Medals'])
data.sort_values(by = 'No of Gold Medals',ascending = False).head().plot(x = 'Sport' , y = 'No of Gold Medals' ,kind = 'bar' , figsize = (3,5))
```

```python
data.iloc[0]
```

### **Which Sport has more no of medals so far (TOP 5)**

```python
sport_list = df.Sport.unique()
data1 = []
for sport in sport_list:
  filt = (df['Sport'] == sport) & ((df['Medal'] == 'Gold') | (df['Medal'] == 'Silver') | (df['Medal'] == 'Bronze'))
  data1.append((sport,len(df[filt])))
data1 = pd.DataFrame(data1 , columns = ['Sport','No of Medals'])
data1.sort_values(by = 'No of Medals',ascending = False).head().plot(x = 'Sport' , y = 'No of Medals' , kind = 'bar' , figsize = (3,5))
```

```python
data1.iloc[0]
```

### **Which Player has won  more no of Medals (TOP 5)**

```python
Athlete_list = df.Athlete.unique()
data2 = []
for Athlete in Athlete_list:
  filt = (df['Athlete'] == Athlete) & ((df['Medal'] == 'Gold') | (df['Medal'] == 'Silver') | (df['Medal'] == 'Bronze'))
  data2.append((Athlete,len(df[filt])))
data2
```

```python
df2 = pd.DataFrame(data2 ,columns=['Athlete','No of medals'])
data2 = df2.sort_values(by = 'No of medals' , ascending = False).head()
data2.plot(x = 'Athlete' , y = 'No of medals' , kind = 'bar',figsize = (3,5))
```

```python
data2.iloc[0]
```

### **Which Player has won more Gold Medals (TOP 5)**

```python
Athlete_list = df.Athlete.unique()
data3 = []
for Athlete in Athlete_list:
  filt = (df['Athlete'] == Athlete) & (df['Medal'] == 'Gold')
  data3.append([Athlete , len(df[filt])])
```

```python
df3 = pd.DataFrame(data3 , columns= ['Athlete','No of Gold Medals'])
df3.sort_values(by = 'No of Gold Medals' ,ascending = False).head().plot(x = 'Athlete' , y = 'No of Gold Medals' ,kind ='bar',figsize = (3,5))
```

```python
df3.sort_values(by = 'No of Gold Medals' ,ascending = False).iloc[0]
```

### **Which Sport is having most Female Gold Medalists ? (TOP5)**

```python
data4=[]
for sport in df.Sport.unique():
  filt = ((df['Sport'] == sport) & (df['Gender'] == 'Women') & (df['Medal'] == 'Gold'))
  data4.append([sport,len(df[filt])])
data4
pd.DataFrame(data4,columns=['Sport','No of Female Gold Medalists']).sort_values(by = 'No of Female Gold Medalists',ascending = False).head().plot(x='Sport',y='No of Female Gold Medalists',kind = 'bar',figsize = (5,5))
```
